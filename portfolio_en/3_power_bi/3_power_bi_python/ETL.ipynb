```python
# FInding NaN Values

des = df.describe()
des.loc['Nan'] = df.isna().sum()
des.loc['%Nan'] = (df.isna().mean())*100
des
```

```python
# Copy Df

df2 = df.copy()
df2.head()
# Date formating and new date columns

df2['order_date'] = pd.to_datetime(df2['order_date'],format='%d-%m-%Y')
df2['order_date_month'] = pd.to_datetime(df2['order_date']).dt.month
df2['order_time'] = pd.to_datetime(df2['order_time']).dt.hour

# Drop innecesary columns 



df2.info()
```

```python
df3 = df2.drop('pizza_name_id',axis=1)

# Splitting ingrediets by column
df3['ingredient1'] = df3['pizza_ingredients'].str.split(',',expand=True)[0]
df3['ingredient2'] = df3['pizza_ingredients'].str.split(',',expand=True)[1]
df3['ingredient3'] = df3['pizza_ingredients'].str.split(',',expand=True)[2]


```
